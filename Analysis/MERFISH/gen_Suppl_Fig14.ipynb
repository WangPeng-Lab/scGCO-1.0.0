```python
fig,ax=plt.subplots(1,4,figsize=(20,4))
for i,geneID in enumerate(pattern_conserved_df_exp.index):
    smooth_factor=result_df.loc[geneID,"smooth_factor"]
    exp =  data_norm.loc[:,geneID]
    exp=(log1p(exp)).values
    newLabels = result_df.loc[geneID][7:]
    p=result_df.loc[geneID,"p_value"]
    fdr=result_df.loc[geneID,"fdr"]
    titleText = geneID + ' ' + 'FDR = ' + str("{:.2e}".format(fdr))
    subplot_voronoi_boundary(geneID, locs, exp,  newLabels, fdr,ax=ax[i],fdr=True,set_title=titleText,fontsize=16) 
    ax[i].set_yticks([])
    ax[i].set_xticks([])
    ax[i].axis("off")
fig.savefig("../../figures//Suppl_Figure14f_MERFISH_4_SV_genes.pdf",format="pdf",bbox_inches="tight")
```

```python
# compare the three sets of predictions to tissue structure
# step 1
# create three vects for tissue structure

#cluster_gene=pattern_conserved_df_exp.index
cluster_gene=pattern_conserved_df_exp.index
exp =  data_norm.iloc[:,1]
exp=(log1p(exp)).values
cellGraph = create_graph_with_weight(locs, exp) 
X=log1p(data_norm.loc[:, cluster_gene])
# X=data_norm.loc[:, cluster_gene]
kmeans=KMeans(n_clusters=2,random_state=0).fit(X)
hmrf_tissue_labels = cut_graph_profile(cellGraph, kmeans.labels_, unary_scale_factor=100, 
                      smooth_factor=50)


fig,ax=plt.subplots(dpi=180)
#image=("../data/HE-MOB-breast-cancer/HE_Rep11_MOB.jpg")
#add_HE_image(image,ax)

colors = ['red',"green"]
#colors = ["C" + str(i) for i in range(2)]
# ax.scatter(locs[:,0], locs[:,1], c=hmrf_tissue5_labels, cmap=matplotlib.colors.ListedColormap(colors),s=30)
ax.scatter(locs[:,0], locs[:,1], c=hmrf_tissue_labels, cmap=matplotlib.colors.ListedColormap(colors),s=20)
points=locs
p_dist = cdist(points, points)    
p_dist[p_dist == 0] = np.max(p_dist, axis = 0)[0]
norm_dist = np.mean(np.min(p_dist, axis = 0))

# find points at edge, add three layers of new points 
x_min = np.min(points, axis = 0)[0] - 3*norm_dist
y_min = np.min(points, axis = 0)[1] - 3*norm_dist
x_max = np.max(points, axis = 0)[0] + 3*norm_dist
y_max = np.max(points, axis = 0)[1] + 3*norm_dist

ax.set_xlim(x_min + norm_dist, x_max - norm_dist)
ax.set_ylim(y_min + norm_dist, y_max - norm_dist)
ax.set_title('MERFISH_kmeans2_hmrfSF50')
plt.savefig("../../figures//Suppl_Figure14e_MERFISH_kmeans2_hmrfSF50.pdf",format="pdf",dpi=180, bbox_inches="tight")
plt.show()
```
