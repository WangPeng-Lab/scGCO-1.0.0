```python
# compare the three sets of predictions to tissue structure
# step 1
# create three vects for tissue structure
# cluster_gene=hvg_genes
DESeq2_only=(set(deseq2_df.g.values)-set(pattern_conserved_df_exp.index.values))&set(data_norm.columns)
scGCO_only=set(pattern_conserved_df_exp.index.values)-set(deseq2_df.g.values)
common=set(pattern_conserved_df_exp.index.values)&set(deseq2_df.g.values)
DESeq2=set(deseq2_df.g.values)&set(data_norm.columns)
scGCO=set(pattern_conserved_df_exp.index.values)
```

```python
def creat_tissue_mat(locs,data_norm,gene_list,num_clusters=5,cluster_size=10):
    gene_subset=gene_list  
    exp =  data_norm.iloc[:,1]
    exp=(log1p(exp)).values
    cellGraph = create_graph_with_weight(locs, exp) 
    X=log1p(data_norm.loc[:, gene_subset])
    tissue_mat_list = list()
    for  num_cluster in range(2,num_clusters+1):
        kmeans=KMeans(n_clusters=num_cluster,random_state=0).fit(X)
        kmeans_labels=kmeans.labels_
        hmrf_labels = cut_graph_profile(cellGraph, kmeans.labels_, unary_scale_factor=100, 
                              smooth_factor=15)
        uniq, count = np.unique(kmeans_labels, return_counts = True)

        for tm_index in np.arange(len(uniq)):
            temp_vec = np.zeros(kmeans_labels.shape[0])
            temp_vec[np.where(kmeans_labels == uniq[tm_index])[0]] = 1
            temp_vec_inv = abs(1- temp_vec)
           
            if max(sum(temp_vec_inv), sum(temp_vec))/min(sum(temp_vec_inv), sum(temp_vec)) < 1.2:
                if np.sum(temp_vec_inv )>cluster_size:
                    tissue_mat_list.append(temp_vec_inv)        
            if np.sum(temp_vec)>cluster_size and np.sum(temp_vec) < kmeans_labels.shape[0]/2:
                tissue_mat_list.append(temp_vec)
    tissue_mat = np.array(tissue_mat_list)
    print(tissue_mat.shape)
    return tissue_mat
tissue_mat=creat_tissue_mat(locs,data_norm,pattern_conserved_df.index,num_clusters=5,cluster_size=10)
```

```python
scGCO_only_df, missed_gene = generate_target_shape_amtrix_for_hamming(data_norm, 
                                                    result_df, scGCO_only)
DESeq2_only_df, missed_gene = generate_target_shape_amtrix_for_hamming(data_norm, 
                                                    result_df, DESeq2_only)
common_df, missed_gene = generate_target_shape_amtrix_for_hamming(data_norm, 
                                                    result_df, common)
# known_genes_df,missed_gene= generate_target_shape_amtrix_for_hamming(data_norm,result_df,known_gene)

#create dummy graph cuts result for missed genes and add to target_df
#this is mainly for DESeq2-only gene
scGCO_only_hamming_df = computer_norm_hamming_to_tissue(tissue_mat, scGCO_only_df, result_df)
DESeq2_only_hamming_df = computer_norm_hamming_to_tissue(tissue_mat, DESeq2_only_df, result_df)
common_hamming_df = computer_norm_hamming_to_tissue(tissue_mat, common_df, result_df)
```
