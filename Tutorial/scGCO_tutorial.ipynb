# Tutorial with Rep11 of MOB 
## This is a step-by-step instruction on running the main functionalities of scGCO.
* <font size=4><mark>Step 1-6:</mark> perform genome-scale identification of spatially variably genes. </font>
* <font size=4><mark>Step 7-8:</mark> visualize and save identified spatial variably genes.</font>
* <font size=4><mark>Step 9:</mark> perform graph cuts on a single genes to visualize its spatial patterns.</font>



## **Step 1:**

### **Read in raw data and perform standard normalization.**

````python
j=11
unary_scale_factor=100
label_cost=10
algorithm='expansion'
ff = '../data/Raw_data/MOB-breast-cancer/Rep' + str(j) +  '_MOB_count_matrix-1.tsv'
locs,data=read_spatial_expression(ff,sep='\t')


data_norm = normalize_count_cellranger(data)
print('Rep{}_processing: {}'.format(j,data_norm.shape))
````

## **Step 2:**  

### **Estimate smooth factor that will minimize false positive rates** 

* <font size=4> **Recommended parameters for Mouse olfactor bulb:** </font> 

    * <font size=3> ```start_sf = 20```</font>

    * <font size=3> ```fdr_cutoff = 0.01```</font>

````python

from sklearn.utils import shuffle

data_norm_rand = shuffle(data_norm)
smooth_factor = estimate_smooth_factor(locs, data_norm_rand,start_sf = 20, fdr_cutoff=0.01)
print('Rep{}: '.format(j),smooth_factor)
````

## **Step 3:**  


### **Run the main scGCO function to identify genes with a non-random spatial variability**

````python
sf=30
result_df=identify_spatial_genes(locs,data_norm,smooth_factor=sf)
````

<br><font size=4>We perform to write and save scGCO output results with <mark><font color=red>**write_result_to_csv**</font></mark> functions for cross-platform.<br><br> Meanwhile, When reread these results we should use <mark><font color=red>**read_result_to_dataframe**</font></mark> functions. <font>

````python
write_result_to_csv(result_df,'../data/MouseOB/scGCO_results//Rep{}_results_df'.format(j))
````

````python
result_df=read_result_to_dataframe('../data/MouseOB/scGCO_results/Rep11_result_df.csv')
print(result_df.shape)
````

## **Step 4:**   

### **Select genes with significant spatial non-random patterns using a specific fdr cutoff.** 

* <font size=4>default: 0.01 <br> 
select genes demonstrating significant spatial variability</font>

````python
fdr_cutoff=0.01
fdr_df=result_df.loc[result_df.fdr<fdr_cutoff,]
print(fdr_df.shape)
````

## **Step 5:**   

### **Select genes with significantly conserved spatial patterns.**


<font size=4>At first, <mark><font color=red >simulate_hamming</font></mark>  function can estimate hamming cutoff with a specific p-value,then suggestes ***0.001 or  

0.0001***, by which <mark><font color=red >cutoff</font></mark>  parameters are to control the similarity threshold.  

<br>
<br>Then, we run <mark><font color=red >identify_pattern_conserved_genes_iteration</font></mark>  function to actually filter genes by conservation of spatial  


patterns. Here, If you want automatically learn cell clusters from the input data,please
set **fixed_k=None**; If the number of clusters (K) is known from priori knowledge, set **fixed_k=K**.  <font> 


````python
## select genes with significantly conserved sptail patterns

_,hamming_df=simulate_hamming(locs,data_norm,fdr_df,cutoff=0.0001)
hamming_cutoff=round(hamming_df,2)
print('Rep{} {}'.format(j,hamming_cutoff))
    
````

````python
fdr_opt=fdr_df.sort_values(by='fdr')

## our estimate hamming was 0.76.

hamming_cutoff=0.76
seg_max=data_norm.shape[0]*0.6 

pattern_df,tissue_mat,targe_df,hamming_df=identify_pattern_conserved_genes_iteration(
    locs, data_norm, fdr_opt,
                            similarity_cutoff=hamming_cutoff,cluster_size=5,smooth_factor=10, 
                            perplexity=30,fixed_k=None,seg_min=8,seg_max=seg_max)

print('Rep{}_step2{}'.format(j,pattern_df.shape))
````

````python
## visualize the learned conserved spatial domains
print('Rep{}_tissue{}'.format(j,tissue_mat.shape))


image='../data/Raw_data//HE-MOB-breast-cancer/HE_Rep{}_MOB.jpg'.format(j)
colors=['green','blue']
title='Rep11'
plot_tissue_pattern(locs,data_norm,tissue_mat,image,colors,title,nrows=5,ncols=5)

plt.show()
````

````python

step2= read_result_to_dataframe('../data/MouseOB/scGCO_results//Rep{}_pattern_df.csv'.format(j))
print(step2.shape)
````

## **Step 9:** 

<br><font size=4>**Perform graph cuts for a single gene.**<font>

````python
# You can also analyze one gene of interest

geneID='Apod' # Lets use Nrgn as an example
unary_scale_factor = 100 # scale factor for unary energy, default value works well

# set smooth factor to 20; 
# use bigger smooth_factor to get less segments
# use small smooth_factor to get more segments
smooth_factor=20 

ff = '../data/Raw_data/MOB-breast-cancer/Rep11_MOB_count_matrix-1.tsv' 
# read in spatial gene expression data
locs, data = read_spatial_expression(ff,sep='\t')

# normalize gene expression
data_norm = normalize_count_cellranger(data)

# select Nrgn's expression
exp =  data_norm.loc[:,geneID]

# log transform
exp=(log1p(exp)).values

# create graph representation of spatial coordinates of cells
cellGraph = create_graph_with_weight(locs, exp)

# do graph cut
newLabels, gmm = cut_graph_general(cellGraph, exp, unary_scale_factor, smooth_factor)
# calculate p values
p, node, com = compute_p_CSR(locs,newLabels, gmm, exp, cellGraph)

# Visualize graph cut results
plot_voronoi_boundary(geneID, locs, exp,  newLabels, min(p)) 

# save the graph cut results to pdf
pdf_voronoi_boundary(geneID, locs, exp, newLabels, min(p), 
                     '../figures//{}.pdf'.format(geneID),
                    point_size=0)
````
